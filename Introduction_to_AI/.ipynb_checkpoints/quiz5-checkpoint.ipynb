```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('data2.txt')
x = data.iloc[:,[0,1]].to_numpy()
y = data.iloc[:,[2]].to_numpy()

bias = np.ones((118,1))
x = np.hstack((bias,x)) 
x.shape
```

```python
max_iter = 100000
learning_rate = 0.001

w = np.array([[0., 0., 0.]]).T
p = 1/(1 + np.exp(-np.dot(x,w)))
gradient = np.array([[0., 0., 0.]]).T
ce = [];
for iteration in range(0,max_iter):
    
    p = 1/(1 + np.exp(-np.dot(x,w)))
    err = 0;
    for i in range(0,x.shape[0]):
        a = float(y[i,0])
        err += a*np.log(p[i,0])+(1-a)*np.log(1-p[i,0])
    err /= x.shape[0]  
    ce.append(err)
    #ce.append(-err)
    for i in range(0,x.shape[0]):
        gradient += x[i,:].reshape(1,x[0,:].shape[0]).T * (p[i,0]-float(y[i,0]))   
    gradient /= x.shape[0]  
    w = w - learning_rate*gradient
x_axis = range(1,max_iter+1)
plt.plot(x_axis,ce)
plt.show()
```

```python
predicted = p = 1/(1 + np.exp(-np.dot(x,w)))
correct = predicted[:,0] >= 0.5
correct = correct.sum()
print('Accuracy(%) : \n{0}'.format(100*correct/x.shape[0]))
```

# code for polynomial 6 (p4,p5)

```python
basis = np.zeros((118,27))

for i in range(1,6+1):
    for j in range(0,i+1):
        for k in range(0,x.shape[0]):
            # general solution for sequence of difference
            col = int(((i*(i+1)/2))-1)+j 
            basis[k,col] = float((x[k,0]**(i-j)))*float((x[k,1]**j))
basis = np.hstack((bias,basis)) 
```

```python
basis = pd.DataFrame(basis)
display(basis)
basis=basis.values
```

```python
max_iter = 10000
learning_rate = 0.001
lambda1 = 1

w = np.zeros((28,1))
p = 1/(1 + np.exp(-np.dot(basis,w)))
gradient = np.zeros((28,1))
for iteration in range(0,max_iter):
    
    p = 1/(1 + np.exp(-1*np.dot(basis,w)))
    for i in range(0,basis.shape[0]):
        gradient += basis[i,:].reshape(28,1) * (p[i,0]-float(y[i,0]))
        
    temp = w
    temp *= lambda1
    temp[0,0] = 0
    gradient += temp
    gradient 
    
    w = w - learning_rate*gradient/basis.shape[0]
```

```python
def PolyFeatures(x1, x2,degree):
    out = [1]
    for i in range(1,degree+1):
        for j in range(0,i+1):
            out.append( float((x1**(i-j)))*float((x2**j)))
    return out
```

```python
u = np.linspace(-1, 1.5, 50)
v = np.linspace(-1, 1.5, 50)
z = np.zeros((len(u), len(v)))
for i in range(len(u)):
    for j in range(len(v)):
         z[i,j] = np.dot(PolyFeatures(u[i],v[j],6),w)
```
