```python
import numpy as np
from matplotlib import pyplot as plt
x = np.array([
    [1, 1, 0],
    [1, 2, 1],
    [1, 3, 2],
    [1, 2, 6]
])

y = np.array([[1, 0, 2, 1]]).T

P = np.dot(x.T,y)/4
R = np.dot(x.T,x)/4
D = np.dot(y.T,y)/4

# print('x : \n{0}'.format(x)); print('\n')
# print('y : \n{0}'.format(y)); print('\n')
print('D : \n{0}'.format(D))
# print('P : \n{0}'.format(P)); print('\n')
# print('R : \n{0}'.format(R))
```

```python
max_iter = 1000
learning_rate = 0.001
w = np.array([[0, 0, 0]]).T
err = [];
for iteration in range(0,max_iter):
    v = np.dot(R,w)
    w = w + learning_rate*(-2*(v-P))
    a_1 = np.dot(P.T,w);
    a_2 = np.dot(w.T,R)
    a_3 = np.dot(a_2,w)
    mse = 0.5*D - a_1 + 0.5*a_3 
#     print(mse)
    err.append(float(mse))
# print('err : \n{0}'.format(err))    
x_axis = range(1,max_iter+1)
plt.plot(x_axis,err)
plt.show()
print(mse)
print('Predicted y :',w.sum())
```

```python
pi = np.array([
    [1, 1, 0, 0, 1, 0],
    [1, 2, 1, 2, 4, 1],
    [1, 3, 2, 6, 9, 4],
    [1, 2, 6, 12, 4, 36]
])

P2 = np.dot(pi.T,y)/4
R2 = np.dot(pi.T,pi)/4

max_iter = 1000
learning_rate = 0.001
w2 = np.array([[0, 0, 0, 0, 0, 0]]).T
err2 = [];
for iteration in range(0,max_iter):
    v = np.dot(R2,w2)
    w2 = w2 + learning_rate*(-2*(v-P2))
    a_1 = np.dot(P2.T,w2);
    a_2 = np.dot(w2.T,R2)
    a_3 = np.dot(a_2,w2)
    mse = float(0.5*D - a_1 + 0.5*a_3 )
#     print(mse)
    err2.append(mse)
# print('err : \n{0}'.format(err))    
x_axis = range(1,max_iter+1)
plt.plot(x_axis,err2)
plt.show()
print(mse)
print('Predicted y :',w2.sum())
```
