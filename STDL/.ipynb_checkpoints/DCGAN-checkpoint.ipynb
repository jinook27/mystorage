# Define Generator & Discriminator

```python
import torch.nn as nn
import torch.nn.init as init
import torch.nn.functional as F
import torch.optim as optim
from torch.optim import lr_scheduler

# Number of channels in the training images. For color images this is 3
nc = 3

# Size of z latent vector (i.e. size of generator input)
nz = 100

# Size of feature maps in generator
ngf = 64

# Size of feature maps in discriminator
ndf = 64

class Generator(nn.Module):
    def __init__(self):
        super(Generator,self).__init__()
        self.convT1 = nn.ConvTranspose2d(nz,ngf*8,4,1,0,bias=False)
        self.convT2 = nn.ConvTranspose2d(ngf*8,ngf*4,4,2,1,bias=False)
        self.convT3 = nn.ConvTranspose2d(ngf*4,ngf*2,4,2,1,bias=False)
        self.convT4 = nn.ConvTranspose2d(ngf*2,ngf,4,2,1,bias=False)
        self.convT5 = nn.ConvTranspose2d(ngf*2,nc,4,2,1,bias=False)
        
        self.batchn1 = nn.BatchNorm2d(ngf*8)
        self.batchn2 = nn.BatchNorm2d(ngf*4)
        self.batchn3 = nn.BatchNorm2d(ngf*2)
        self.batchn4 = nn.BatchNorm2d(ngf)
        
    def foward(self,x):
        x = F.relu(self.batchn1(self.ConvT1(x)))
        x = F.relu(self.batchn2(self.ConvT2(x)))
        x = F.relu(self.batchn3(self.ConvT3(x)))
        x = F.relu(self.batchn4(self.ConvT4(x)))
        x = ConvT5(x)
        return F.Tanh(x)
    
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator,self).__init__()
        self.conv1 = nn.Conv2d(nc, ndf, 4, 2, 1,bias = False)
        self.conv2 = nn.Conv2d(ndf, ndf*2, 4, 2, 1,bias = False)
        self.conv3 = nn.Conv2d(ndf*2, ndf*4, 4, 2, 1,bias = False)
        self.conv4 = nn.Conv2d(ndf*4, ndf*8, 4, 2, 1,bias = False)
        self.conv5 = nn.Conv2d(ndf*8, 1, 4, 1, 0,bias = False)
        
        self.batchn1 = nn.BatchNorm2d(ndf)
        self.batchn2 = nn.BatchNorm2d(ndf*2)
        self.batchn3 = nn.BatchNorm2d(ndf*4)
        self.batchn4 = nn.BatchNorm2d(ndf*8)
        
    def foward(self,x):
        x = self.batchn1(self.conv1(x))
        x = F.leaky_relu(x,0.2,inplace=True)
        x = self.batchn2(self.conv2(x))
        x = F.leaky_relu(x,0.2,inplace=True)
        x = self.batchn3(self.conv3(x))
        x = F.leaky_relu(x,0.2,inplace=True)
        x = self.batchn4(self.conv4(x))
        x = F.leaky_relu(x,0.2,inplace=True)
        x = conv5(x)
        x = F.Sigmoid(x)
        return x.view(-1,1).squeeze(1)
    
def weights_init(m):
    class_name = m.__class__.__name__
    if class_name.find('conv') !=-1:
        m.weight.data.normal_(0.0,0.02)
        #nn.init.xavier_normal(m.weight.data)
    elif class_name.find('batchn') !=-1:
        m.weight.data.normal_(1.0,0.02)
        #m.bias.data.fill(0)
        torch.nn.init.zeros_(m.bias.data)
        
Generator = Generator()
Generator.apply(weights_init)
Discriminator = Discriminator()
Discriminator.apply(weights_init)
```
